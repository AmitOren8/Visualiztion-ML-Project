```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('md2.csv', index_col='ID')
df.head()
```

```python
df_d = df[['Age', 'High_Education', 'Relationship','Income', 'Childrenhome', 'NumDealsPurchases']]
df_d.head()
```

```python
X = df_d.drop('NumDealsPurchases', axis=1)
y = df_d['NumDealsPurchases']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
rf = RandomForestRegressor(n_estimators = 100, random_state = 0)
rf.fit(X_train, y_train)
```

```python
result = rf.score(X_test, y_test)
print("Accuracy: %.3f%%" % (result*100.0))
```
