```python
import pandas as pd
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('md.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df1 = df[['Age', 'Education', 'Marital_Status', 'Income']]
df1.head()
```

```python
df1.dtypes
```

```python
df1.isnull().sum()
```

```python
df1= pd.get_dummies(df1)
df1.head()
```

```python
imputer = KNNImputer(n_neighbors=3)
imputed = imputer.fit_transform(df1)
df_imputed = pd.DataFrame(imputed, columns=df1.columns)
```

```python
df_imputed.isnull().sum()
```

```python
df_imputed['Income']
```

```python
df['Income'] = df_imputed['Income']
```

```python
df.isnull().sum()
```

```python
df.to_csv('md1.csv')
```
